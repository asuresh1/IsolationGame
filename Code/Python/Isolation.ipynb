```python
class isolation(object):
    def __init__(self, shape, players):
        """Initialize Isolation Game"""
        # A game symbol legend for reference 
        # and future expansion
        self.legend = {'unblocked':0,
                       'blocked':1}
        # The shape of the n-dimensional board
        self.shape = shape
        # Initialize this n-dimensional board
        self.initBoard()
        # Initialize the player dictionary
        self.initPlayers(players)
        # Render the initialize boards display
        self.showBoard()
        
    def initBoard(self):
        """Initialize Board"""
        # Initialize the board with unblocked zeros
        self.board = np.zeros(self.shape, dtype=int)
    
    def initPlayers(self, players):
        """Initialize Players"""
        # Initialize the dictionary of players
        # for convenient access and bookkeeping
        self.players = {}
        for playerID, player in enumerate(players):
            # Make sure no player has number symbol
            # as within the symbol legend
            playerID += 2
            # Each player has the player ID
            # for rendering position on the board,
            # a player object to hold the AI algorithm,
            # and the current location for bookkeeping
            player = {'playerID': playerID,
                      'player': player,
                      'spot': None}
            self.players[playerID] = player
    
    def showBoard(self,raw=False):
        """Show Board"""
        if raw:
            # For debug
            print(self.board)
        else:
            # Print a pretty bored
            self.showColumnLable()
            for i, row in enumerate(self.board):
                rowStr = str("%2d" % (i+1)) + str(row)
                print(rowStr)
    
    def showColumnLable(self):
        """Show Column Alpha Index"""
        columnLable = '   '
        for col in range(self.shape[1]):
            quot, rem = divmod(col, 26)
            if quot:
                columnLable += chr(quot-1 + ord('A')) + chr(rem + ord('A'))
            else:
                columnLable += chr(rem + ord('A'))
            columnLable += ' '
        print(columnLable)
    
    def getMove(self, player):
        """Get Player Move"""
        # Get all the available moves
        moves = self.allMoves(player['spot'])
        try:
            # Half the player choose a move
            move = player['player'].getMoveAI(self.board, self.legend, player['playerID'], player['spot'], moves)
        except:
            # If the player code failed
            # make sure to specify who
            print('Fail Whale Player:', player['playerID'])
            raise Exception("Player Failer")
        if not self.isMove(move, moves):
            # If the player made an invalid move
            # make sure you specify who
            print('Bad Move Player:', player['playerID'])
            raise Exception("Player Move is Invalid")
        return move
    
    def allMoves(self, spot):
        """Finds all the moves for a player"""
        if spot is None:
            # If the player is not already on the board
            # then return all the possible unblocked spots
            moves = np.vstack(np.where(self.board==self.legend['unblocked'])).T
            return moves
        else:
            # Use recursion to search from the starting spot
            # across all the horizontal verticals and diagonals
            # to list all of the possible moves
            moves  = self.isValid(spot,[-1,-1])
            moves += self.isValid(spot,[-1, 0])
            moves += self.isValid(spot,[-1, 1])
            moves += self.isValid(spot,[ 0, 1])
            moves += self.isValid(spot,[ 1, 1])
            moves += self.isValid(spot,[ 1, 0])
            moves += self.isValid(spot,[ 1,-1])
            moves += self.isValid(spot,[ 0,-1])
            return np.array(moves)
    
    def isValid(self, spot, neighbor):
        """Checks if Neighbor is valid"""
        # Calculate the index for the index
        index = np.array(spot)+np.array(neighbor)
        # Check to make sure it's not negative
        # as we are not allowing wrapping
        if np.array(np.where(index<0)).size:
            return []
        index = tuple(index)
        # Check to see if it's out of bounds
        try:
            value = self.board[index]
        except:
            return []
        # Check to see if it's unblocked
        if value != self.legend['unblocked']:
            return []
        # If it's available, return its index
        # along with the result for the next neighbor
        else:
            return [index] + self.isValid(index, neighbor)
    
    def isMove(self, move, moves):
        """Checks if Move is valid"""
        # Make sure that the move made
        # is among one of the valid moves given
        if move in moves:
            return True
        else:
            return False
    
    def isDead(self, player):
        """Checks if Player is Dead"""
        moves = self.allMoves(player['spot'])
        # The number moves left for a given player
        # determines their final fate
        if moves.size:
            return False
        else:
            # If the player has no more moves
            # declare their horrible death
            # and remove them from the remaining contenders
            print 'Player', player['playerID'], 'died a horrible death'
            self.players.pop(player['playerID'])
            return True
        
    def playMove(self, player):
        """Let player move on the board"""
        # Alter the board by moving the player
        # and leave a blocked cell in their wake
        playerID = player['playerID']
        newSpot = self.getMove(player)
        oldSpot = self.players[playerID]['spot']
        self.players[playerID]['spot'] = newSpot
        self.board[tuple(newSpot)] = playerID
        # if the player has no previous location
        # a cell on the board cannot be blocked
        if oldSpot is not None:
            self.board[tuple(oldSpot)] = self.legend['blocked']
        
    def isDone(self):
        """Checks if Game is done"""
        # Check the termination condition
        # when only one player remains
        return len(self.players) <= 1
        
    def play(self):
        # Perpetually playing turns
        # until the game is done
        while not self.isDone():
            # Grab the list of competing players
            players = self.players.values()
            # Let each player have their turn
            for player in players:
                # For every turn
                # check if we are done
                if self.isDone():
                    break
                # Check to make sure the player isn't dead
                if not self.isDead(player):
                    print 'Player', player['playerID'], 'move'
                    # Let the player make a move
                    self.playMove(player)
                    # Render the new board for display
                    self.showBoard()
        # The winner will be the only remaining
        # player within the list of players
        winner  = self.players.values()[0]
        print'And the Winner is:', winner['playerID']
        try:
            # Let the winner have a bit of fun
            winner['player'].taunt()
        except:
            print "Woot!"
```
